# 🛒 Market Basket Analysis using Apyori
This notebook uses the Apyori package to generate association rules.

```
import pandas as pd
from apyori import apriori
```

```
# Sample dataset
dataset = [
    ['Milk', 'Bread', 'Butter'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread'],
    ['Milk', 'Butter'],
    ['Bread'],
    ['Milk', 'Bread', 'Butter'],
    ['Eggs', 'Milk'],
    ['Eggs', 'Bread']
]
```

```
# Apply Apriori Algorithm
rules = apriori(dataset, min_support=0.3, min_confidence=0.6, min_lift=1.0, min_length=2)
```

```
# Display results
results = list(rules)
for item in results:
    pair = item[0]
    items = [x for x in pair]
    print('Rule:', items)
    print('Support:', item[1])
    print('Confidence:', item[2][0][2])
    print('Lift:', item[2][0][3])
    print('------------------------')
```

## 📊 Conclusion
The Apyori package helps identify strong association rules between items.
